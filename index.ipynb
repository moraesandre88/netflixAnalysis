### Creating a df

```python
df_1 = pd.DataFrame(df.groupby('country').show_id.count())
df_1['latitude'] = pd.Series(dtype='float64')
df_1['longitude'] = pd.Series(dtype='float64')
display(df_1)
```

```python
df_1 = df_1.reset_index().rename(columns={'show_id':'releases'})
display(df_1)
```

```python
for i in df_1.index:
    for j in coord.index:
        if df_1.country[i].lower() == coord.country[j].lower():
            df_1.latitude[i] = coord.latitude[j]
            df_1.longitude[i] = coord.longitude[j]
        else:
            pass

display(df_1)
```

```python
df_1.isnull().sum()
```

```python
df_1 = df_1.dropna()
df_1.isnull().sum()
```

```python
display(df_1)
```

```python
aux = pd.DataFrame()

df_1['often_rated'] = pd.Series(dtype='object')
df_1['percent_movie'] = pd.Series(dtype='float64')
df_1['percent_tv_show'] = pd.Series(dtype='float64')
df_1['percent_global_ranking'] = pd.Series(dtype='float64')
df_1['global_ranking'] = df_1['releases'].rank(method='min', ascending=False)

for i in df_1.index:

    aux = pd.DataFrame(df.loc[df['country'] == df_1.country[i], ['rating']].mode())
    df_1.often_rated[i] = random.choice(aux.rating.values)
    df_1.percent_global_ranking[i] = round(df_1.releases[i]/df_1['releases'].sum() * 100, 2)
    
    aux = pd.DataFrame(df.loc[(df['country'] == df_1.country[i])].type.value_counts())
    
    for j in aux.index:
        if j == 'Movie':
            df_1.percent_movie[i] = round((aux.type[j]/aux.type.sum()) * 100, 2)
        else:
            df_1.percent_tv_show[i] = round((aux.type[j]/aux.type.sum()) * 100, 2)

df_1 = df_1.fillna(0.00)



display(df_1)
```

```python
my_map = folium.Map(
    location=[13,16],
    zoom_start=2
)

for i in df_1.index:

    plt.figure()
    plt.title('Produced content (%)')
    data = [df_1.percent_movie[i], df_1.percent_tv_show[i]]
    labels = ['Movie', 'Tv Show']
    colors = sns.color_palette('pastel')[0:5]
    plt.pie(data, labels = labels, colors = colors, autopct='%.2f%%')
    name = 'figure{}.png'.format(i)
    plt.savefig(name)
    plt.close()
    
    encoded = base64.b64encode(open(name, 'rb').read())

    html = (
        """
        <h1>{}</h1><br>
        >Releases: {} ({}% of all produced)<br>
        >Global ranking: {}Â°<br>
        >Most commom rating: {}<br>
        <p></p>
        <img src="data:image/png;base64,{}">
        """
    ).format(df_1.country[i], df_1.releases[i], df_1.percent_global_ranking[i], int(df_1.global_ranking[i]), df_1.often_rated[i], encoded.decode('UTF-8'))

    iframe = branca.element.IFrame(html=html, width=470, height=470)
    popup = folium.Popup(iframe, max_width=2650)

    folium.Marker(
        location=[df_1.latitude[i],df_1.longitude[i]],
        tooltip=df_1.country[i],
        popup=popup,
    ).add_to(my_map)

my_map
```
