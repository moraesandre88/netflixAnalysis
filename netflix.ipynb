### Creating a df with the coordinates to plot using Folium

```python
df_1 = pd.DataFrame(df.groupby('country').show_id.count())
df_1['latitude'] = pd.Series()
df_1['longitude'] = pd.Series()
display(df_1)
```

```python
df_1 = df_1.reset_index().rename(columns={'show_id':'releases'})
display(df_1)
```

```python
#len(np.unique(df_1.country))
```

```python
#len(np.unique(df.country))
```

```python
for i in df_1.index:
    for j in coord.index:
        if df_1.country[i].lower() == coord.country[j].lower():
            df_1.latitude[i] = coord.latitude[j]
            df_1.longitude[i] = coord.longitude[j]
        else:
            pass

display(df_1)
```

```python
df_1.isnull().sum()
```

```python
df_1 = df_1.dropna()
df_1.isnull().sum()
```

```python
display(df_1)
```

```python
#df_1.releases.sum()
```

```python
aux_1 = pd.DataFrame()
aux_2 = pd.DataFrame()
aux_3 = pd.DataFrame()
df_1['often_rated'] = pd.Series()
df_1['often_type'] = pd.Series()
df_1['percent_movie'] = pd.Series()
df_1['percent_tv_show'] = pd.Series()

for i in df_1.index:
    aux_1 = df.loc[df['country'] == df_1.country[i], ['rating']].mode()
    aux_2 = df.loc[df['country'] == df_1.country[i], ['type']].mode()
    aux_3 = df.loc[(df['country'] == df_1.country[i])]
    aux_3 = pd.DataFrame(aux_3.type.value_counts())
    df_1.often_rated[i] = random.choice(aux_1.rating.values)
    df_1.often_type[i] = random.choice(aux_2.type.values)
    for j in aux_3.index:
        if j == 'Movie':
            df_1.percent_movie[i] = round((aux_3.type[j]/aux_3.type.sum()) * 100, 2)
        else:
            df_1.percent_tv_show[i] = round((aux_3.type[j]/aux_3.type.sum()) * 100, 2)

df_1 = df_1.fillna(0.00)

display(df_1)
```

```python
my_map = folium.Map(
    location=[13,16],
    zoom_start=2
)

for i in df_1.index:
    text = (
        """ 
        <h1>{}</h1><br>
        >Number of releases: {}<br>
        >Most commom type: {}<br>
        >Most commom ratings: {}
        """
    ).format(df_1.country[i], df_1.releases[i], df_1.often_type[i], df_1.often_rated[i])

    iframe = branca.element.IFrame(html=text, width=300, height=300)
    popup = folium.Popup(iframe, max_width=2500)

    folium.Marker(
        location=[df_1.latitude[i],df_1.longitude[i]],
        tooltip=df_1.country[i],
        popup=popup
    ).add_to(my_map)
my_map
```
